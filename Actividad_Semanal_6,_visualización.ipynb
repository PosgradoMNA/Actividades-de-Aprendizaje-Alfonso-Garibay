# 7.- Reduce las dimensiones con PCA, si consideras necesario.

```python
#Estandarizar los datos
pcs = PCA()
pcs.fit(preprocessing.scale(Data3.iloc[:, 0:14].dropna(axis=0))) 
```

```python
#Intentamos encontrar cuantos componenetes son necesarios para llegar al 70% de la información. 
#Para esto imprimimos la desviación estandar, la proporción de la varianzam y la propoción acumulada.

pcsSummary_df = pd.DataFrame({'Standard deviation': np.sqrt(pcs.explained_variance_),
                             'Proportion of variance': pcs.explained_variance_ratio_,
                             'Cumulative proportion': np.cumsum(pcs.explained_variance_ratio_)
                             })

pcsSummary_df = pcsSummary_df.transpose()
pcsSummary_df.columns = ['PC{}'.format(i) for i in range(1, len(pcsSummary_df.columns) + 1)]
pcsSummary_df.round(4) 
```

**Basandonos en la proporción acumulada podemos ver que necesitamos 5 componentes principales que explican 70% de la información.**



```python
pcsComponents_df = pd.DataFrame(pcs.components_.transpose(), 
                                columns = pcsSummary_df.columns,
                                index = Data3.iloc[:, 0:14].columns
                                )
pcsComponents_df.iloc[:,:5]
```

```python
#Convertir pcsComponents en valores absolutos para hacer mas facil su revision
abspcsComponents=pd.DataFrame(data=pcsComponents_df.abs())
```

```python
#Analisis de las dos variables más importantes para el calculo de cada componente

print("Las dos variables más importantes para PCA 1 son: \n", abspcsComponents.PC1.nlargest(2),"\n")
print("Las dos variables más importantes para PCA 2 son: \n", abspcsComponents.PC2.nlargest(2),"\n")
print("Las dos variables más importantes para PCA 3 son: \n", abspcsComponents.PC3.nlargest(2),"\n")
print("Las dos variables más importantes para PCA 4 son: \n", abspcsComponents.PC4.nlargest(2),"\n")
print("Las dos variables más importantes para PCA 5 son: \n", abspcsComponents.PC5.nlargest(2),"\n")
```

```python
#Analisis de las dos variables menos importantes para el calculo de cada componente

print("Las dos variables menos importantes para PCA 1 son: \n", abspcsComponents.PC1.nsmallest(2),"\n")
print("Las dos variables menos importantes para PCA 2 son: \n", abspcsComponents.PC2.nsmallest(2),"\n")
print("Las dos variables menos importantes para PCA 3 son: \n", abspcsComponents.PC3.nsmallest(2),"\n")
print("Las dos variables menos importantes para PCA 4 son: \n", abspcsComponents.PC4.nsmallest(2),"\n")
print("Las dos variables menos importantes para PCA 5 son: \n", abspcsComponents.PC5.nsmallest(2),"\n")
```
